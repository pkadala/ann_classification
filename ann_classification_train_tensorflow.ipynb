```python
import sklearn
import pandas as pd 

from sklearn.preprocessing import StandardScaler, LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
```

## Load the data

```python
data = pd.read_csv("Churn_Modelling.csv")
data.head()
```

## Remove unwanted columns

```python
data = data.drop(columns=['RowNumber','CustomerId', 'Surname'], axis=1)
data
```

## Encode categorical variables

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_gender = LabelEncoder()
data['Gender'] = labelencoder_gender.fit_transform(data['Gender'])
data
```

```python
## For categories more than one like Geography ONE HOT ENCODING must be used

from sklearn.preprocessing import OneHotEncoder
onehotencoder_geo = OneHotEncoder()
ohe_encoded_geo = onehotencoder_geo.fit_transform(data[['Geography']]).toarray()
ohe_encoded_geo
```

```python
## Create a seperate data frame for one hot encoded values 
# so we can append to the original data frame and remove the unencoded geography colum
print(onehotencoder_geo.get_feature_names_out(['Geography']))

one_geo_df = pd.DataFrame(ohe_encoded_geo, columns=onehotencoder_geo.get_feature_names_out(['Geography']))
one_geo_df
```

```python
## Print the original data frame
data
```

```python
# Remove the geography from the original data frame
data = data.drop('Geography',axis=1)
```

```python
##  Add the one hot encoded data frame to the original data frame

data = pd.concat([data, one_geo_df], axis=1)
data
```

```python
## Save the encoders and the scalars to the file so we can re-use it at later point 
# for the same problem domain
import pickle

with (open('labelencoder_gender.pkl' , 'wb')) as file:
    pickle.dump(labelencoder_gender, file)

with (open('onehotencoder_geo.pkl', 'wb')) as file:
    pickle.dump(onehotencoder_geo, file)    
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler

## Split the data set to independent and dependent variables
X = data.drop(['Exited'], axis=1)
y = data['Exited']

## Split the dataset into test train set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=52)

## Scale these features
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

## ANN Implementation


1. Import necessary classes
2. Build a model  (and check the summary)
3. Compile a model (by passing optimizers, loss and metrics)
4. Setup the tensorboard and early stopping
5. Train the model

```python
## Import necessary tensorflow and keras objects

import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping
```

```python
## Build a model

model = Sequential ([
    Dense(64, activation='relu', input_shape=X_train[1].shape),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.summary()
```

```python
## Compile the model
import tensorflow
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy

adam_optimizer = Adam(learning_rate=0.01)
bcross_loss = BinaryCrossentropy()

model.compile(optimizer=adam_optimizer, loss= bcross_loss, metrics=['accuracy'])
```

```python
## Initialize the callbacks for training the model
import datetime
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping

log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)

earlystopping_callback = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
## Train the model

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    callbacks=[tensorboard_callback, earlystopping_callback]
)
```

```python
## Save the model

model.save('ann_coassification_model.keras')
```
